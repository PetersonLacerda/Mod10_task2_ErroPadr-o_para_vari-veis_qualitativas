````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
previsao_de_renda = pd.read_csv('/content/previsao_de_renda.csv')
````

````python
previsao_de_renda.head()
````

````python
ax = sns.pointplot(x="posse_de_imovel", y="renda",
                   data=previsao_de_renda, dodge=True, ci = 90)
````

````python
plt.figure(figsize=(10, 6))
ax = sns.pointplot(x="educacao", y="renda", hue = 'posse_de_imovel',
                   data=previsao_de_renda, dodge=True, ci = 90)

plt.legend(loc = 'lower center', bbox_to_anchor=(0.5, -.30),ncol = 3)
````

````python
plt.figure(figsize=(10, 6))
ax = sns.pointplot(x="tipo_renda", y="renda", hue = 'posse_de_imovel',
                   data=previsao_de_renda, dodge=True, ci = 90)

plt.legend(loc = 'lower center', bbox_to_anchor=(0.5, -.30),ncol = 3)
````

Para 'posse_de_imovel', vemos que a renda para quem possui tem uma variação menor e um valor médio maior. 

Quando abrimos a visão e inserimos a variável 'ensino', os níveis considerados superiores, superiores incompletos e pós-graduação, tem um número de pessoas que possuem casa maior.

Para tipo de renda, as variações de quem possui ou não imóvel tem uma variação menor, exeto para a classe considerada Bolsista.

````python
ax = sns.pointplot(x="posse_de_veiculo", y="renda",
                   data=previsao_de_renda, dodge=True, ci = 90)
````

````python
plt.figure(figsize=(10, 6))
ax = sns.pointplot(x="educacao", y="renda", hue = 'posse_de_veiculo',
                   data=previsao_de_renda, dodge=True, ci = 95)

plt.legend(loc = 'lower center', bbox_to_anchor=(0.5, -.30),ncol = 3)
````

````python
plt.figure(figsize=(10, 6))
ax = sns.pointplot(x="tipo_renda", y="renda", hue = 'posse_de_veiculo',
                   data=previsao_de_renda, dodge=True, ci = 95)

plt.legend(loc = 'lower center', bbox_to_anchor=(0.5, -.30),ncol = 3)
````

Para a variável 'posse_de_veiculo' a renda das pessoas que possuem veículo é substancialmente maior.

Ao dividirmos por variáveis de 'educacao', vemos comportamentos distintos para as classes 'Secundário', 'Superior' e Superior incompleto, vemos as pessoas com veículo, com uma renda substancialmente maior.
A renda da classe primária para quem tem ou não veículo tem a sua similaridade.
Já para a classe de pós graduação, vemos a renda de quem não tem veículo substancialmente maior.

Empregando a classe 'tipo_renda', as pessoas que possuem veículo tem uma renda maior em todas as classes de tipo de renda.
